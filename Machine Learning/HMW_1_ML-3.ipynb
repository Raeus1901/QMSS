3. Calculate the average of the mpg column for all cars within each category of the cyl column.

4. Create a histogram using the mpg column


```python
%matplotlib inline
from matplotlib import pyplot as plt

df['mpg'].hist()
plt.xlabel('Miles/(US) gallon')
plt.ylabel('Count')

# On a s.d, the graph can be represented through a single line of coding thus saving time => df.hist(column='mpg')
```

5. Choose two variables in the data and create a scatterplot.


```python
plt.scatter(df['mpg'], df['wt'])
plt.title("Miles/(US) gallon vs. Weight (1000 lbs)")
plt.xlabel("Miles/(US) gallon")
plt.ylabel("Weight (1000 lbs)")
plt.show()

#The lighter the vehicle, the longer the potential distance it can reach per galon.
```
